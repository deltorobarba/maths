1. Ausgangslage und konzeptioneller Rahmen

Der klassische Fall ist die glatte Mannigfaltigkeit $M$ mit ihrer **kommutativen** Algebra $A = C^\infty(M)$ der glatten Funktionen. Topologische Invarianten erh├ñlt man hier ├╝ber die **de Rham-Kohomologie** $H^*_{dR}(M)$, also durch Integration von Differentialformen $\omega \in \Omega^n(M)$ ├╝ber $M$.

Connes' Leitfrage (1981ÔÇô1985): *Was bleibt, wenn man die Punkte und die Kommutativit├ñt verliert?* Sein Geniestreich besteht in der Erkenntnis, dass die algebraische Essenz eines Integrals ÔÇö eine multilineare Form $\varphi(f_0, df_1, \dots, df_n)$ mit **zyklischer Symmetrie** (via Stokes: $\int d(\dots)=0$) und **graduierter Antisymmetrie** in den $df_i$ ÔÇö ein autonomes Konzept ist, das *keine* Mannigfaltigkeit voraussetzt. ├£brig bleibt eine Kohomologie f├╝r jede assoziative Algebra, die im kommutativen, glatten Fall wieder $H^*_{dR}$ liefert. Der konzeptionelle Schritt: **Integration ├╝ber Mannigfaltigkeit ÔåÆ zyklische Spur auf Algebra**.

2. De Rham-Kohomologie (das kommutative Referenzobjekt)

**Definition.** F├╝r eine glatte Mannigfaltigkeit $M$ sei $\Omega^k(M)$ der Raum der glatten $k$-Formen. Die ├ñu├ƒere Ableitung $d: \Omega^k(M) \to \Omega^{k+1}(M)$ ist lokal definiert durch

$$d\bigl(f\, dx_{i_1}\wedge\dots\wedge dx_{i_k}\bigr) = \sum_j \frac{\partial f}{\partial x_j}\, dx_j \wedge dx_{i_1} \wedge \dots \wedge dx_{i_k},$$

und die Kohomologiegruppen sind

$$H^k_{dR}(M) = \frac{\ker(d:\Omega^k\to\Omega^{k+1})}{\operatorname{im}(d:\Omega^{k-1}\to\Omega^k)}.$$

Das Keilprodukt $\wedge$ ist antikommutativ ($dx\wedge dy = -dy\wedge dx$) und erf├╝llt die **graduierte Leibniz-Regel** f├╝r $\alpha\in\Omega^p, \beta\in\Omega^q$:

$$d(\alpha\wedge\beta) = d\alpha\wedge\beta + (-1)^p \alpha\wedge d\beta.$$

**Rechenbeispiel.** Im $\mathbb{R}^2$ mit $\alpha = x\,dy$, $\beta = y\,dx$:

Direkt: $\alpha\wedge\beta = -xy\,(dx\wedge dy)$, und $d(-xy\,dx\wedge dy)=0$, weil $dx\wedge dx=0$ und $dy\wedge dy=0$.

Via Leibniz ($p=1$): $d\alpha = dx\wedge dy$, $d\beta = -dx\wedge dy$, und

$$d\alpha\wedge\beta - \alpha\wedge d\beta = y(dx\wedge dy\wedge dx) + x(dy\wedge dx\wedge dy) = 0.$$

Beide Wege liefern $0$ ÔÇö die Struktur ist konsistent.

3. Hochschild-Komplex (das Fundament)

Sei $A$ eine assoziative, unitale $\mathbb{C}$-Algebra. Der Raum der $n$-Koketten ist

$$C^n(A) := \operatorname{Hom}_{\mathbb{C}}(A^{\otimes(n+1)}, \mathbb{C}),$$

also der Raum aller multilinearen Funktionale $\varphi: A^{n+1}\to\mathbb{C}$. Der **Hochschild-Korandoperator** $b: C^n(A)\to C^{n+1}(A)$ ist

$$(b\varphi)(a_0,\dots,a_{n+1}) = \sum_{i=0}^n (-1)^i \varphi(a_0,\dots,a_i a_{i+1},\dots,a_{n+1}) + (-1)^{n+1}\varphi(a_{n+1}a_0, a_1,\dots,a_n).$$

Es gilt $b^2 = 0$. Die Hochschild-Kohomologie $HH^n(A) := H^n(C^*(A), b)$ ist das nichtkommutative Analogon zu $\Omega^n(M)$.

**Konkret in Grad 0:** Eine 0-Kokette $\varphi: A\to\mathbb{C}$ erf├╝llt $b\varphi=0$ genau dann, wenn $\varphi(a_0 a_1) = \varphi(a_1 a_0)$ f├╝r alle $a_0, a_1\in A$ ÔÇö das hei├ƒt: *Eine 0-Kokokette im Kern von $b$ ist genau eine Spur auf $A$.* Spuren ├╝bernehmen in der nichtkommutativen Welt die Rolle der Integration von 0-Formen.

4. Zyklische Kohomologie via zyklische Symmetrie

Der zyklische Permutationsoperator $\lambda: C^n(A)\to C^n(A)$ ist

$$(\lambda\varphi)(a_0,\dots,a_n) = (-1)^n \varphi(a_n, a_0,\dots,a_{n-1}).$$

Eine Kokette hei├ƒt **zyklisch**, falls $\lambda\varphi = \varphi$. Der Unterraum

$$C^n_\lambda(A) := \{\varphi\in C^n(A): \lambda\varphi = \varphi\}$$

ist unter $b$ invariant (Connes' Beobachtung: $b(C^n_\lambda)\subset C^{n+1}_\lambda$). Damit ist $(C^*_\lambda(A), b)$ ein Teilkomplex, und

$$\boxed{\;HC^n(A) := H^n(C^*_\lambda(A), b) = \frac{\ker\bigl(b: C^n_\lambda(A)\to C^{n+1}_\lambda(A)\bigr)}{\operatorname{im}\bigl(b: C^{n-1}_\lambda(A)\to C^n_\lambda(A)\bigr)}\;}$$

ist die **zyklische Kohomologie** von $A$.

5. Der $(b,B)$-Bikomplex (die rechnerisch m├ñchtige Beschreibung)

Die $\lambda$-Definition ist anschaulich, in der Praxis aber unhandlich ÔÇö die zyklische Bedingung ist extrem restriktiv, und Quantenalgebren erlauben selten eine explizite Isolierung von $C^n_\lambda$. Connes' zweite Beschreibung umgeht das.

**Der Connes-Korandoperator** $B: C^n(A)\to C^{n-1}(A)$ wird als Komposition definiert:

$$B = N \circ s \circ (1-\lambda), \quad N = \sum_{i=0}^{n-1}\lambda^i \text{ (Norm)}, \quad (s\varphi)(a_0,\dots,a_{n-1}) = \varphi(1, a_0,\dots,a_{n-1}).$$

Explizit:

$$(B\varphi)(a_0,\dots,a_{n-1}) = \sum_{i=0}^{n-1}(-1)^{(n-1)i}\,\varphi(1, a_i,\dots,a_{n-1}, a_0,\dots,a_{i-1}).$$

Es gelten die fundamentalen Identit├ñten

$$b^2 = 0, \qquad B^2 = 0, \qquad bB + Bb = 0.$$

Damit lebt die zyklische Kohomologie auf einem zweidimensionalen Gitter, dem **$(b,B)$-Bikomplex** mit horizontalem $b$ (Grad +1) und vertikalem $B$ (Grad ÔêÆ1):

$$\begin{array}{ccccc}
\vdots & & \vdots & & \\
\uparrow B & & \uparrow B & & \\
C^2(A) & \xrightarrow{b} & C^3(A) & \xrightarrow{b} & \cdots \\
\uparrow B & & \uparrow B & & \\
C^1(A) & \xrightarrow{b} & C^2(A) & \xrightarrow{b} & \cdots \\
\uparrow B & & \uparrow B & & \\
C^0(A) & \xrightarrow{b} & C^1(A) & \xrightarrow{b} & \cdots
\end{array}$$

Die zyklische Kohomologie ist die Kohomologie des Totalkomplexes ÔÇö ├ñquivalent zur $\lambda$-Definition (**Connes' Theorem**), aber rechnerisch deutlich handhabbarer, weil man im *unrestringierten* Raum $C^n(A)$ arbeitet.

**Rechenbeispiel ($B$ in Aktion, $n=1$).** F├╝r eine 1-Kokette $\varphi\in C^1(A)$ kollabieren die drei Zahnr├ñder ($(1-\lambda)$, $s$, $N$) zu

$$(B\varphi)(a_0) = \varphi(1, a_0) + \varphi(a_0, 1).$$

Sei $A = M_2(\mathbb{C})$ und $\psi(M_1, M_2) = \operatorname{Tr}(M_1 M_2)$. Dann

$$(B\psi)(a_0) = \operatorname{Tr}(I\cdot a_0) + \operatorname{Tr}(a_0\cdot I) = 2\operatorname{Tr}(a_0).$$

$B$ hat aus der 1-Kokette die fundamentale 0-dimensionale Struktur (die Spur) extrahiert.

**Rechenbeispiel ($bB+Bb=0$, kommutativer Fall).** F├╝r $A=\mathbb{C}[x]$ und die 1-Kokette $\phi(f_0,f_1) = f_0(0)f_1'(0)$ ergibt sich $(B\phi)(f_0) = f_0'(0)$, und beide Wege liefern Null: $b(B\phi)(f_0,f_1) = (f_0 f_1)'(0) - (f_1 f_0)'(0) = 0$ (wegen Kommutativit├ñt), und $(b\phi)(f_0,f_1,f_2) = 0$ nach Produktregel-Aufrechnung. Also $(bB+Bb)\phi = 0$.

**Warum braucht man $B$ ├╝berhaupt?** Drei Gr├╝nde: (i) Rechnen in $C^n_\lambda$ ist bei komplexen Algebren praktisch unm├Âglich; im Bikomplex arbeitet man mit allen Koketten. (ii) $B$ ist das *echte* geometrische Analogon der ├ñu├ƒeren Ableitung $d$, w├ñhrend $b$ eine algebraische Assoziativit├ñtspr├╝fung ist. (iii) Ohne den Bikomplex existiert der Periodizit├ñtsoperator $S$ nicht ÔÇö und damit keine periodische zyklische Kohomologie und keine Br├╝cke zur K-Theorie.

6. Die SBI-Sequenz (Connes-Periodizit├ñt)

Hochschild- und zyklische Kohomologie h├ñngen ├╝ber die exakte Sequenz

$$\cdots \to HH^n(A) \xrightarrow{I} HC^n(A) \xrightarrow{S} HC^{n+2}(A) \xrightarrow{B} HH^{n+1}(A) \to \cdots$$

zusammen. Die drei Akteure: $I$ vergisst die zyklische Symmetrie (Hochschild ist ÔÇ×gr├Â├ƒer"); $S$ ist der **Periodizit├ñtsoperator** und erh├Âht den Grad um 2 ÔÇö das Herzst├╝ck der ganzen Theorie; $B$ ist der Connes-Korand als verbindender Homomorphismus zur├╝ck in die Hochschild-Welt.

7. Periodische zyklische Kohomologie $HP$

Da $S$ die Dimension stets um 2 erh├Âht, ist der direkte Limes wohldefiniert:

$$\boxed{\;HP^n(A) := \varinjlim\Bigl(HC^n(A) \xrightarrow{S} HC^{n+2}(A) \xrightarrow{S} HC^{n+4}(A) \xrightarrow{S} \cdots\Bigr).\;}$$

$HP$ ist **2-periodisch**: $HP^n(A) \cong HP^{n+2}(A)$. Es bleiben also nur zwei Gruppen, $HP^{\mathrm{even}}(A)$ und $HP^{\mathrm{odd}}(A)$. Sie sind der nat├╝rliche Empf├ñnger des **Chern-Charakters** und das eigentliche Rechenwerkzeug der nichtkommutativen Geometrie: Sie filtern das ÔÇ×Rauschen" der gew├Âhnlichen $HC^*$ heraus und liefern die puren, robusten topologischen Invarianten von Quantenr├ñumen (z. B. Hall-Leitf├ñhigkeit, topologische Isolatoren).

8. Interpretation der Ergebnisse: Werkzeugpr├╝fung vs. Messung

Eine wichtige Unterscheidung, die oft Verwirrung stiftet: Es gibt zwei Ebenen, auf denen eine Zahl (Null oder nicht) auftauchen kann.

**Ebene 1 ÔÇö Werkzeugpr├╝fung ($b\varphi = 0$?):** Hier pr├╝ft man, ob $\varphi$ ├╝berhaupt ein Kozyklus ist, also als Messger├ñt zugelassen wird. $b\varphi=0$ bedeutet: g├╝ltiger Kozyklus. $b\varphi\neq 0$: nicht zugelassen.

**Ebene 2 ÔÇö eigentliche Messung ($\varphi(a_0,\dots,a_n)=?$):** Mit einem g├╝ltigen Kozyklus $\varphi$ misst man konkrete Tupel. Ergebnis 0 hei├ƒt: auf *dieser* Route keine topologische Aktivit├ñt. Ergebnis $\neq 0$ hei├ƒt: hier ist eine geometrische Verklebung, ein Hindernis, eine umrundete ÔÇ×Achse".

**Quantenbeispiel.** Mit $\psi(a_0, a_1) = \tfrac{i}{2}\operatorname{Tr}(a_0[H, a_1])$ und $H=\sigma_z$: Diese Funktion ist ein zyklischer 1-Kozyklus ($b\psi=0$ automatisch wegen Spureigenschaften). Die Messungen ergeben $\psi(\sigma_x,\sigma_x)=0$ (keine Aktivit├ñt), aber $\psi(\sigma_y,\sigma_x)=-2$ ÔÇö der Hamiltonoperator verklebt die Richtungen $\sigma_x$ und $\sigma_y$ aktiv miteinander (Pr├ñzession).

9. Cup-Produkt und Ringstruktur

In der Homologie kann man Ketten addieren, aber nicht multiplizieren. In der Kohomologie hingegen sind Koketten *Funktionale*, deren Werte sich multiplizieren lassen ÔÇö das ergibt das **Cup-Produkt** $\smile$. F├╝r $\phi\in C^p(X)$, $\psi\in C^q(X)$ ist $\phi\smile\psi\in C^{p+q}(X)$ definiert durch

$$(\phi\smile\psi)([v_0,\dots,v_{p+q}]) = \phi([v_0,\dots,v_p]) \cdot \psi([v_p,\dots,v_{p+q}]),$$

also ÔÇ×Vorderseite mal R├╝ckseite" des Simplex. Damit wird $H^*(X) = \bigoplus_n H^n(X)$ ein **graduiert kommutativer Ring**:

$$\alpha\smile\beta = (-1)^{pq}(\beta\smile\alpha) \quad \text{f├╝r } \alpha\in H^p, \beta\in H^q.$$

**Rechenbeispiel ($\Delta^2$ mit Ecken $v_0, v_1, v_2$).** Setze $\phi(e_{01})=1$ (sonst 0), $\psi(e_{12})=1$ (sonst 0). Dann

$$(\phi\smile\psi)([v_0,v_1,v_2]) = \phi([v_0,v_1])\cdot\psi([v_1,v_2]) = 1\cdot 1 = 1,$$

aber $(\psi\smile\phi)([v_0,v_1,v_2]) = \psi([v_0,v_1])\cdot\phi([v_1,v_2]) = 0\cdot 0 = 0$.

Auf Kokettenebene also nicht kommutativ; erst nach ├£bergang zu Kohomologieklassen wird das Produkt graduiert kommutativ. Geometrisch misst das Cup-Produkt das *Hintereinanderkleben* von Eigenschaften: erst $\phi$-Bedingung, dann direkt $\psi$-Bedingung.

Im ├£bergang zu Differentialformen wird $\smile$ zum **Keilprodukt** $\wedge$, und der Korandoperator $\delta$ zum ├ñu├ƒeren Differential $d$. Connes' Definition der zyklischen Kohomologie reproduziert im kommutativen, glatten Fall genau diese Struktur.

10. Kovariant vs. Kontravariant ÔÇö die Pfeil-Asymmetrie

Sei $\psi: A\to B$ ein Algebrenhomomorphismus.

**Zyklische Homologie $HC_*$ ist kovariant.** Ketten der Form $a_0\otimes\dots\otimes a_n$ lassen sich vorw├ñrts schieben:

$$\psi_*(a_0\otimes\dots\otimes a_n) = \psi(a_0)\otimes\dots\otimes\psi(a_n), \qquad \psi_*: HC_n(A)\to HC_n(B).$$

**Zyklische Kohomologie $HC^*$ ist kontravariant.** Koketten sind Funktionale $\varphi: B^{n+1}\to\mathbb{C}$, und der induzierte Pfeil l├ñuft *r├╝ckw├ñrts*:

$$(\psi^*\varphi)(a_0,\dots,a_n) = \varphi(\psi(a_0),\dots,\psi(a_n)), \qquad \psi^*: HC^n(B)\to HC^n(A).$$

**Warum die Asymmetrie?** Die Komposition zwingt die Richtung. Aus $C_n(X)\xrightarrow{f_*} C_n(Y)\xrightarrow{\varphi}\mathbb{Z}$ entsteht eine Kokette auf $X$, nicht auf $Y$. Da der Zielraum eines Funktionals starr der Skalark├Ârper ist, kann der Pfeil am Ende nicht umgebogen werden ÔÇö Funktionale (Messger├ñte) lassen sich nur *zur├╝ckziehen* (Pullback), nicht vorw├ñrtsschieben.

**Rechenbeispiel.** Sei $\psi: \mathbb{C}[x]\to\mathbb{C}$, $\psi(p) = p(0)$ (Auswertung am Ursprung). Ein lineares Funktional $\varphi(z) = cz$ auf $\mathbb{C}$ wird zu

$$(\psi^*\varphi)(p) = \varphi(\psi(p)) = c\cdot p(0).$$

Aus einem Messger├ñt auf dem Ziel $B = \mathbb{C}$ ist ein Messger├ñt auf dem Ursprung $A = \mathbb{C}[x]$ geworden ÔÇö der Pfeil wurde umgedreht.

**Quanten-Br├╝ckenschlag.** Diese Dualit├ñt ist exakt der Unterschied zwischen Schr├Âdinger- und Heisenberg-Bild. Schr├Âdinger ist kovariant ($\rho_t = U_t\rho_0 U_t^\dagger$, Zust├ñnde werden vorw├ñrts geschoben). Heisenberg ist kontravariant ($A_t = U_t^\dagger A_0 U_t$, Observablen werden zur├╝ckgezogen ÔÇö daher das adjungierte $U_t^\dagger$ vorne, das genauso auftaucht wie bei kontravarianten Tensoren in der ART).

11. HKR-Theorem als Br├╝cke

Das **Hochschild-Kostant-Rosenberg-Theorem** schl├ñgt die Br├╝cke zur├╝ck zum klassischen Fall: F├╝r glatte kommutative Algebren $A = C^\infty(M)$ gilt

$$HH_n(A) \cong \Omega^n(M),$$

und die zyklische Kohomologie reproduziert in diesem Fall die de Rham-Kohomologie. Connes' Definition ist also so geschnitten, dass sie im kommutativen, glatten Fall die klassische Theorie *exakt* zur├╝ckgewinnt ÔÇö bleibt aber starr genug, um auf wilden, nichtkommutativen Quantenalgebren sinnvolle topologische Aussagen zu treffen.

1. Multilineare Abbildung
* Bilinearformen, das euklidische Skalarprodukt. Es nimmt zwei Vektoren aus dem kartesischen Produkt und macht eine Zahl daraus: $B(a, b) = a \cdot b \in \mathbb{R}$. Das ist eine Funktion, noch keine Algebra von Elementen.

2. Tensoralgebra $T(V)$:
* bilden das Tensorprodukt $a \otimes b$. Dies ist weder ein Skalar noch ein reiner Fl├ñcheninhalt. Es ist eine Dyade, ein $3 \times 3$-Matrix-artiges Objekt, das als linearer Operator fungieren kann. Es codiert sowohl den Teil, der parallel ist, als auch den, der senkrecht steht. Hier ist alles erlaubt, nichts wird vereinfacht. Ein Tensor 2. Stufe $v \otimes w$ ist einfach nur ein abstraktes Paar.
* Warum haben wir das Tensorprodukt $V \otimes W$ ├╝berhaupt erfunden? Um multilineare Abbildungen in gew├Âhnliche *lineare* Abbildungen zu verwandeln! In der Kategorientheorie gibt es einen wunderbaren Isomorphismus. Die Menge aller bilinearen Abbildungen $B: V \times V \to \mathbb{R}$ ist isomorph zur Menge der rein linearen Abbildungen $L: V \otimes V \to \mathbb{R}$. Das bedeutet: Jede Bilinearform ist streng genommen ein Funktional auf dem Tensorprodukt! Das Tensorprodukt wurde exakt daf├╝r konstruiert, die multilineare Algebra zu algebraisieren.

3. ├äu├ƒeren Algebra $\Lambda(V)$:
* rein topologisch. Wir bilden das Keilprodukt $a \wedge b$. Das innere Produkt (die L├ñngen-Information) wird hier rigoros vernichtet. Es bleibt *nur* der reine **Bivektor**, der das orientierte Parallelogramm beschreibt, das $a$ und $b$ aufspannen. $a \wedge a = 0$. Wir nehmen die Tensoralgebra und erzwingen die Regel $v \otimes v = 0$. Dadurch entsteht das **Keilprodukt** ($\wedge$). Hier werden abstrakte Bivektoren ($v \wedge w$) geboren. Sie sind antisymmetrische Tensoren. Diese Algebra hat keine Metrik, sie beschreibt nur Volumina und Schnittpunkte.

4. Geometrischen Algebra $Cl(V, q)$:
* geometrisch/metrisch. Wir bilden das geometrische Produkt $ab$. Hier f├╝gt sich alles zusammen! Die Geometrische Algebra nimmt die Bilinearform (1) und den abstrakten Bivektor aus der ├äu├ƒeren Algebra (3) und summiert sie auf: $ab = B(a,b) + (a \wedge b)$. Das Ergebnis ist ein "Multivektor", der gleichzeitig einen Skalarteil und einen Bivektorteil besitzt. * **Bivektoren und Keilprodukte in der Geometrischen Algebra**: Weil die Geometrische Algebra als Vektorraum *isomorph* (strukturgleich) zur ├äu├ƒeren Algebra ist!
