  * **Linear Transformations**: Generated by linear or quadratic functions ($\le 2^{\text{nd}}$ degree) of fundamental operators $\hat{Q}, \hat{P}$, or $\hat{X}$ and $\hat{Z}$ (modulo 𝑑). Not orthogonal and form overcomplete basis. Eigenstates of annihilation operator $\hat{a}$.

<font color="blue">**Periodische, zyklischen Kohomologie**

<font color="blue">*Wichtige Zusatznotizen*

* Messung von Veränderungen: Geometrie eine Differentialform = in der Algebra einen Kozyklus

1. Ausgangslage und konzeptioneller Rahmen

Der klassische Fall ist die glatte Mannigfaltigkeit $M$ mit ihrer **kommutativen** Algebra $A = C^\infty(M)$ der glatten Funktionen. Topologische Invarianten erhält man hier über die **de Rham-Kohomologie** $H^*_{dR}(M)$, also durch Integration von Differentialformen $\omega \in \Omega^n(M)$ über $M$.

Connes' Leitfrage (1981–1985): *Was bleibt, wenn man die Punkte und die Kommutativität verliert?* Sein Geniestreich besteht in der Erkenntnis, dass die algebraische Essenz eines Integrals — eine multilineare Form $\varphi(f_0, df_1, \dots, df_n)$ mit **zyklischer Symmetrie** (via Stokes: $\int d(\dots)=0$) und **graduierter Antisymmetrie** in den $df_i$ — ein autonomes Konzept ist, das *keine* Mannigfaltigkeit voraussetzt. Übrig bleibt eine Kohomologie für jede assoziative Algebra, die im kommutativen, glatten Fall wieder $H^*_{dR}$ liefert. Der konzeptionelle Schritt: **Integration über Mannigfaltigkeit → zyklische Spur auf Algebra**.

2. De Rham-Kohomologie (das kommutative Referenzobjekt)

**Definition.** Für eine glatte Mannigfaltigkeit $M$ sei $\Omega^k(M)$ der Raum der glatten $k$-Formen. Die äußere Ableitung $d: \Omega^k(M) \to \Omega^{k+1}(M)$ ist lokal definiert durch

$$d\bigl(f\, dx_{i_1}\wedge\dots\wedge dx_{i_k}\bigr) = \sum_j \frac{\partial f}{\partial x_j}\, dx_j \wedge dx_{i_1} \wedge \dots \wedge dx_{i_k},$$

und die Kohomologiegruppen sind

$$H^k_{dR}(M) = \frac{\ker(d:\Omega^k\to\Omega^{k+1})}{\operatorname{im}(d:\Omega^{k-1}\to\Omega^k)}.$$

Das Keilprodukt $\wedge$ ist antikommutativ ($dx\wedge dy = -dy\wedge dx$) und erfüllt die **graduierte Leibniz-Regel** für $\alpha\in\Omega^p, \beta\in\Omega^q$:

$$d(\alpha\wedge\beta) = d\alpha\wedge\beta + (-1)^p \alpha\wedge d\beta.$$

**Rechenbeispiel.** Im $\mathbb{R}^2$ mit $\alpha = x\,dy$, $\beta = y\,dx$:

Direkt: $\alpha\wedge\beta = -xy\,(dx\wedge dy)$, und $d(-xy\,dx\wedge dy)=0$, weil $dx\wedge dx=0$ und $dy\wedge dy=0$.

Via Leibniz ($p=1$): $d\alpha = dx\wedge dy$, $d\beta = -dx\wedge dy$, und

$$d\alpha\wedge\beta - \alpha\wedge d\beta = y(dx\wedge dy\wedge dx) + x(dy\wedge dx\wedge dy) = 0.$$

Beide Wege liefern $0$ — die Struktur ist konsistent.

3. Hochschild-Komplex (das Fundament)

Sei $A$ eine assoziative, unitale $\mathbb{C}$-Algebra. Der Raum der $n$-Koketten ist

$$C^n(A) := \operatorname{Hom}_{\mathbb{C}}(A^{\otimes(n+1)}, \mathbb{C}),$$

also der Raum aller multilinearen Funktionale $\varphi: A^{n+1}\to\mathbb{C}$. Der **Hochschild-Korandoperator** $b: C^n(A)\to C^{n+1}(A)$ ist

$$(b\varphi)(a_0,\dots,a_{n+1}) = \sum_{i=0}^n (-1)^i \varphi(a_0,\dots,a_i a_{i+1},\dots,a_{n+1}) + (-1)^{n+1}\varphi(a_{n+1}a_0, a_1,\dots,a_n).$$

Es gilt $b^2 = 0$. Die Hochschild-Kohomologie $HH^n(A) := H^n(C^*(A), b)$ ist das nichtkommutative Analogon zu $\Omega^n(M)$.

**Konkret in Grad 0:** Eine 0-Kokette $\varphi: A\to\mathbb{C}$ erfüllt $b\varphi=0$ genau dann, wenn $\varphi(a_0 a_1) = \varphi(a_1 a_0)$ für alle $a_0, a_1\in A$ — das heißt: *Eine 0-Kokokette im Kern von $b$ ist genau eine Spur auf $A$.* Spuren übernehmen in der nichtkommutativen Welt die Rolle der Integration von 0-Formen.

4. Zyklische Kohomologie via zyklische Symmetrie

Der zyklische Permutationsoperator $\lambda: C^n(A)\to C^n(A)$ ist

$$(\lambda\varphi)(a_0,\dots,a_n) = (-1)^n \varphi(a_n, a_0,\dots,a_{n-1}).$$

Eine Kokette heißt **zyklisch**, falls $\lambda\varphi = \varphi$. Der Unterraum

$$C^n_\lambda(A) := \{\varphi\in C^n(A): \lambda\varphi = \varphi\}$$

ist unter $b$ invariant (Connes' Beobachtung: $b(C^n_\lambda)\subset C^{n+1}_\lambda$). Damit ist $(C^*_\lambda(A), b)$ ein Teilkomplex, und

$$\boxed{\;HC^n(A) := H^n(C^*_\lambda(A), b) = \frac{\ker\bigl(b: C^n_\lambda(A)\to C^{n+1}_\lambda(A)\bigr)}{\operatorname{im}\bigl(b: C^{n-1}_\lambda(A)\to C^n_\lambda(A)\bigr)}\;}$$

ist die **zyklische Kohomologie** von $A$.

5. Der $(b,B)$-Bikomplex (die rechnerisch mächtige Beschreibung)

Die $\lambda$-Definition ist anschaulich, in der Praxis aber unhandlich — die zyklische Bedingung ist extrem restriktiv, und Quantenalgebren erlauben selten eine explizite Isolierung von $C^n_\lambda$. Connes' zweite Beschreibung umgeht das.

**Der Connes-Korandoperator** $B: C^n(A)\to C^{n-1}(A)$ wird als Komposition definiert:

$$B = N \circ s \circ (1-\lambda), \quad N = \sum_{i=0}^{n-1}\lambda^i \text{ (Norm)}, \quad (s\varphi)(a_0,\dots,a_{n-1}) = \varphi(1, a_0,\dots,a_{n-1}).$$

Explizit:

$$(B\varphi)(a_0,\dots,a_{n-1}) = \sum_{i=0}^{n-1}(-1)^{(n-1)i}\,\varphi(1, a_i,\dots,a_{n-1}, a_0,\dots,a_{i-1}).$$

Es gelten die fundamentalen Identitäten

$$b^2 = 0, \qquad B^2 = 0, \qquad bB + Bb = 0.$$

Damit lebt die zyklische Kohomologie auf einem zweidimensionalen Gitter, dem **$(b,B)$-Bikomplex** mit horizontalem $b$ (Grad +1) und vertikalem $B$ (Grad −1):

$$\begin{array}{ccccc}
\vdots & & \vdots & & \\
\uparrow B & & \uparrow B & & \\
C^2(A) & \xrightarrow{b} & C^3(A) & \xrightarrow{b} & \cdots \\
\uparrow B & & \uparrow B & & \\
C^1(A) & \xrightarrow{b} & C^2(A) & \xrightarrow{b} & \cdots \\
\uparrow B & & \uparrow B & & \\
C^0(A) & \xrightarrow{b} & C^1(A) & \xrightarrow{b} & \cdots
\end{array}$$

Die zyklische Kohomologie ist die Kohomologie des Totalkomplexes — äquivalent zur $\lambda$-Definition (**Connes' Theorem**), aber rechnerisch deutlich handhabbarer, weil man im *unrestringierten* Raum $C^n(A)$ arbeitet.

**Rechenbeispiel ($B$ in Aktion, $n=1$).** Für eine 1-Kokette $\varphi\in C^1(A)$ kollabieren die drei Zahnräder ($(1-\lambda)$, $s$, $N$) zu

$$(B\varphi)(a_0) = \varphi(1, a_0) + \varphi(a_0, 1).$$

Sei $A = M_2(\mathbb{C})$ und $\psi(M_1, M_2) = \operatorname{Tr}(M_1 M_2)$. Dann

$$(B\psi)(a_0) = \operatorname{Tr}(I\cdot a_0) + \operatorname{Tr}(a_0\cdot I) = 2\operatorname{Tr}(a_0).$$

$B$ hat aus der 1-Kokette die fundamentale 0-dimensionale Struktur (die Spur) extrahiert.

**Rechenbeispiel ($bB+Bb=0$, kommutativer Fall).** Für $A=\mathbb{C}[x]$ und die 1-Kokette $\phi(f_0,f_1) = f_0(0)f_1'(0)$ ergibt sich $(B\phi)(f_0) = f_0'(0)$, und beide Wege liefern Null: $b(B\phi)(f_0,f_1) = (f_0 f_1)'(0) - (f_1 f_0)'(0) = 0$ (wegen Kommutativität), und $(b\phi)(f_0,f_1,f_2) = 0$ nach Produktregel-Aufrechnung. Also $(bB+Bb)\phi = 0$.

**Warum braucht man $B$ überhaupt?** Drei Gründe: (i) Rechnen in $C^n_\lambda$ ist bei komplexen Algebren praktisch unmöglich; im Bikomplex arbeitet man mit allen Koketten. (ii) $B$ ist das *echte* geometrische Analogon der äußeren Ableitung $d$, während $b$ eine algebraische Assoziativitätsprüfung ist. (iii) Ohne den Bikomplex existiert der Periodizitätsoperator $S$ nicht — und damit keine periodische zyklische Kohomologie und keine Brücke zur K-Theorie.

6. Die SBI-Sequenz (Connes-Periodizität)

Hochschild- und zyklische Kohomologie hängen über die exakte Sequenz

$$\cdots \to HH^n(A) \xrightarrow{I} HC^n(A) \xrightarrow{S} HC^{n+2}(A) \xrightarrow{B} HH^{n+1}(A) \to \cdots$$

zusammen. Die drei Akteure: $I$ vergisst die zyklische Symmetrie (Hochschild ist „größer"); $S$ ist der **Periodizitätsoperator** und erhöht den Grad um 2 — das Herzstück der ganzen Theorie; $B$ ist der Connes-Korand als verbindender Homomorphismus zurück in die Hochschild-Welt.

7. Periodische zyklische Kohomologie $HP$

Da $S$ die Dimension stets um 2 erhöht, ist der direkte Limes wohldefiniert:

$$\boxed{\;HP^n(A) := \varinjlim\Bigl(HC^n(A) \xrightarrow{S} HC^{n+2}(A) \xrightarrow{S} HC^{n+4}(A) \xrightarrow{S} \cdots\Bigr).\;}$$

$HP$ ist **2-periodisch**: $HP^n(A) \cong HP^{n+2}(A)$. Es bleiben also nur zwei Gruppen, $HP^{\mathrm{even}}(A)$ und $HP^{\mathrm{odd}}(A)$. Sie sind der natürliche Empfänger des **Chern-Charakters** und das eigentliche Rechenwerkzeug der nichtkommutativen Geometrie: Sie filtern das „Rauschen" der gewöhnlichen $HC^*$ heraus und liefern die puren, robusten topologischen Invarianten von Quantenräumen (z. B. Hall-Leitfähigkeit, topologische Isolatoren).

8. Interpretation der Ergebnisse: Werkzeugprüfung vs. Messung

Eine wichtige Unterscheidung, die oft Verwirrung stiftet: Es gibt zwei Ebenen, auf denen eine Zahl (Null oder nicht) auftauchen kann.

**Ebene 1 — Werkzeugprüfung ($b\varphi = 0$?):** Hier prüft man, ob $\varphi$ überhaupt ein Kozyklus ist, also als Messgerät zugelassen wird. $b\varphi=0$ bedeutet: gültiger Kozyklus. $b\varphi\neq 0$: nicht zugelassen.

**Ebene 2 — eigentliche Messung ($\varphi(a_0,\dots,a_n)=?$):** Mit einem gültigen Kozyklus $\varphi$ misst man konkrete Tupel. Ergebnis 0 heißt: auf *dieser* Route keine topologische Aktivität. Ergebnis $\neq 0$ heißt: hier ist eine geometrische Verklebung, ein Hindernis, eine umrundete „Achse".

**Quantenbeispiel.** Mit $\psi(a_0, a_1) = \tfrac{i}{2}\operatorname{Tr}(a_0[H, a_1])$ und $H=\sigma_z$: Diese Funktion ist ein zyklischer 1-Kozyklus ($b\psi=0$ automatisch wegen Spureigenschaften). Die Messungen ergeben $\psi(\sigma_x,\sigma_x)=0$ (keine Aktivität), aber $\psi(\sigma_y,\sigma_x)=-2$ — der Hamiltonoperator verklebt die Richtungen $\sigma_x$ und $\sigma_y$ aktiv miteinander (Präzession).

9. Cup-Produkt und Ringstruktur

In der Homologie kann man Ketten addieren, aber nicht multiplizieren. In der Kohomologie hingegen sind Koketten *Funktionale*, deren Werte sich multiplizieren lassen — das ergibt das **Cup-Produkt** $\smile$. Für $\phi\in C^p(X)$, $\psi\in C^q(X)$ ist $\phi\smile\psi\in C^{p+q}(X)$ definiert durch

$$(\phi\smile\psi)([v_0,\dots,v_{p+q}]) = \phi([v_0,\dots,v_p]) \cdot \psi([v_p,\dots,v_{p+q}]),$$

also „Vorderseite mal Rückseite" des Simplex. Damit wird $H^*(X) = \bigoplus_n H^n(X)$ ein **graduiert kommutativer Ring**:

$$\alpha\smile\beta = (-1)^{pq}(\beta\smile\alpha) \quad \text{für } \alpha\in H^p, \beta\in H^q.$$

**Rechenbeispiel ($\Delta^2$ mit Ecken $v_0, v_1, v_2$).** Setze $\phi(e_{01})=1$ (sonst 0), $\psi(e_{12})=1$ (sonst 0). Dann

$$(\phi\smile\psi)([v_0,v_1,v_2]) = \phi([v_0,v_1])\cdot\psi([v_1,v_2]) = 1\cdot 1 = 1,$$

aber $(\psi\smile\phi)([v_0,v_1,v_2]) = \psi([v_0,v_1])\cdot\phi([v_1,v_2]) = 0\cdot 0 = 0$.

Auf Kokettenebene also nicht kommutativ; erst nach Übergang zu Kohomologieklassen wird das Produkt graduiert kommutativ. Geometrisch misst das Cup-Produkt das *Hintereinanderkleben* von Eigenschaften: erst $\phi$-Bedingung, dann direkt $\psi$-Bedingung.

Im Übergang zu Differentialformen wird $\smile$ zum **Keilprodukt** $\wedge$, und der Korandoperator $\delta$ zum äußeren Differential $d$. Connes' Definition der zyklischen Kohomologie reproduziert im kommutativen, glatten Fall genau diese Struktur.

10. Kovariant vs. Kontravariant — die Pfeil-Asymmetrie

Sei $\psi: A\to B$ ein Algebrenhomomorphismus.

**Zyklische Homologie $HC_*$ ist kovariant.** Ketten der Form $a_0\otimes\dots\otimes a_n$ lassen sich vorwärts schieben:

$$\psi_*(a_0\otimes\dots\otimes a_n) = \psi(a_0)\otimes\dots\otimes\psi(a_n), \qquad \psi_*: HC_n(A)\to HC_n(B).$$

**Zyklische Kohomologie $HC^*$ ist kontravariant.** Koketten sind Funktionale $\varphi: B^{n+1}\to\mathbb{C}$, und der induzierte Pfeil läuft *rückwärts*:

$$(\psi^*\varphi)(a_0,\dots,a_n) = \varphi(\psi(a_0),\dots,\psi(a_n)), \qquad \psi^*: HC^n(B)\to HC^n(A).$$

**Warum die Asymmetrie?** Die Komposition zwingt die Richtung. Aus $C_n(X)\xrightarrow{f_*} C_n(Y)\xrightarrow{\varphi}\mathbb{Z}$ entsteht eine Kokette auf $X$, nicht auf $Y$. Da der Zielraum eines Funktionals starr der Skalarkörper ist, kann der Pfeil am Ende nicht umgebogen werden — Funktionale (Messgeräte) lassen sich nur *zurückziehen* (Pullback), nicht vorwärtsschieben.

**Rechenbeispiel.** Sei $\psi: \mathbb{C}[x]\to\mathbb{C}$, $\psi(p) = p(0)$ (Auswertung am Ursprung). Ein lineares Funktional $\varphi(z) = cz$ auf $\mathbb{C}$ wird zu

$$(\psi^*\varphi)(p) = \varphi(\psi(p)) = c\cdot p(0).$$

Aus einem Messgerät auf dem Ziel $B = \mathbb{C}$ ist ein Messgerät auf dem Ursprung $A = \mathbb{C}[x]$ geworden — der Pfeil wurde umgedreht.

**Quanten-Brückenschlag.** Diese Dualität ist exakt der Unterschied zwischen Schrödinger- und Heisenberg-Bild. Schrödinger ist kovariant ($\rho_t = U_t\rho_0 U_t^\dagger$, Zustände werden vorwärts geschoben). Heisenberg ist kontravariant ($A_t = U_t^\dagger A_0 U_t$, Observablen werden zurückgezogen — daher das adjungierte $U_t^\dagger$ vorne, das genauso auftaucht wie bei kontravarianten Tensoren in der ART).

11. HKR-Theorem als Brücke

Das **Hochschild-Kostant-Rosenberg-Theorem** schlägt die Brücke zurück zum klassischen Fall: Für glatte kommutative Algebren $A = C^\infty(M)$ gilt

$$HH_n(A) \cong \Omega^n(M),$$

und die zyklische Kohomologie reproduziert in diesem Fall die de Rham-Kohomologie. Connes' Definition ist also so geschnitten, dass sie im kommutativen, glatten Fall die klassische Theorie *exakt* zurückgewinnt — bleibt aber starr genug, um auf wilden, nichtkommutativen Quantenalgebren sinnvolle topologische Aussagen zu treffen.

12. Quick Reference — die Bausteine auf einen Blick

- $A^{\otimes(n+1)}$ — Tensorprodukt; eine Kokette frisst $n+1$ Algebraelemente als getrennte Argumente.
- $C^n(A)$ — $n$-Koketten: multilineare Funktionale $A^{n+1}\to\mathbb{C}$.
- $b: C^n\to C^{n+1}$ — Hochschild-Korand; hebt den Grad.
- $\lambda$ — zyklischer Permutator; bewegt das letzte Argument nach vorn mit Vorzeichen $(-1)^n$.
- $C^n_\lambda$ — Unterraum der zyklisch invarianten Koketten.
- $B: C^n\to C^{n-1}$ — Connes-Korand; senkt den Grad, geometrisches Analogon zu $d$.
- $S: HC^n\to HC^{n+2}$ — Periodizitätsoperator; Schlüssel zur 2-Periodizität.
- $I$ — Inklusion $HH^n\hookrightarrow HC^n$ (vergisst zyklische Symmetrie).
- $HH^n(A)$ — Hochschild, analog zu $\Omega^n(M)$.
- $HC^n(A)$ — zyklische Kohomologie, analog zu $H^n_{dR}(M)$ im kommutativen Fall.
- $HP^{\mathrm{even/odd}}(A)$ — periodische zyklische Kohomologie, Empfänger des Chern-Charakters.


<font color="blue">*Rechenbeispiel: $A = M_n(\mathbb{C})$*

* **$A = M_n(\mathbb{C})$** (wo die Spur als Urgenerator aller zyklischen Klassen erscheint). dann $A = M_n(\mathbb{C})$, in dem die **Spur als Urbild aller zyklischen Kozyklen** sichtbar wird.
* $A = M_n(\mathbb{C})$ - Dieses Beispiel ist viel reicher, weil hier die **Spur** als fundamentale zyklische Klasse explizit erscheint.
1. Setup: Sei $A = M_n(\mathbb{C})$ die Algebra der $n \times n$-Matrizen. Wir suchen einen Erzeuger von $HC^0(A)$.
  * **Definition:** $\;\tau_0 \in C^0(A)$ via $\tau_0(A) := \mathrm{Tr}(A)$ (die Matrix-Spur).
2. Verifikation: $\tau_0$ ist Hochschild-Kozyklus
  * Wir rechnen: $(b\tau_0)(A_0, A_1) = \tau_0(A_0 A_1) - \tau_0(A_1 A_0) = \mathrm{Tr}(A_0 A_1) - \mathrm{Tr}(A_1 A_0).$
  * Die **fundamentale Spureigenschaft** $\mathrm{Tr}(XY) = \mathrm{Tr}(YX)$ liefert sofort $b\tau_0 = 0. \qquad \checkmark$
  * Beobachten Sie: Die Hochschild-Kozyklus-Bedingung für Grad $0$ ist *genau* die Spur-Eigenschaft. Daher gilt allgemein $HC^0(A) = HH^0(A) = \{\text{stetige Spuren auf } A\}.$
3. Verifikation: $\tau_0$ ist zyklisch invariant
  * Für $n = 0$ ist die zyklische Bedingung $(\lambda\tau_0)(A_0) = (-1)^0 \tau_0(A_0) = \tau_0(A_0)$, also automatisch erfüllt. Damit $[\tau_0] \in HC^0(M_n(\mathbb{C})).$
4. Eindeutigkeit: Bestimmung von $HC^0$
  * Da $M_n(\mathbb{C})$ einfach ist und jede Spur ein Vielfaches der Standardspur ist: $HC^0(M_n(\mathbb{C})) = \mathbb{C} \cdot [\tau_0] \cong \mathbb{C}.$
5. **Konstruktion höhergradiger zyklischer Kozyklen**
 * Hier kommt die *Periodizität* konkret zum Tragen. Wir definieren für jedes $k \geq 0$: $\tau_{2k}(A_0, A_1, \dots, A_{2k}) := \mathrm{Tr}(A_0 A_1 \cdots A_{2k}).$
* **Schritt A: Hochschild-Kozyklus**
$$(b\tau_{2k})(A_0,\dots,A_{2k+1}) = \sum_{i=0}^{2k} (-1)^i \,\mathrm{Tr}(A_0 \cdots A_i A_{i+1} \cdots A_{2k+1}) \;+\; (-1)^{2k+1}\, \mathrm{Tr}(A_{2k+1} A_0 \cdots A_{2k}).$$

* Beobachtung 1: In jedem Innenterm bleibt das Produkt $A_0 A_1 \cdots A_{2k+1}$ unverändert (Assoziativität).
Beobachtung 2: Der letzte Term liefert via Spur-Zyklizität $\mathrm{Tr}(A_{2k+1} A_0 \cdots A_{2k}) = \mathrm{Tr}(A_0 \cdots A_{2k+1})$.

* Damit
$$(b\tau_{2k})(A_0,\dots,A_{2k+1}) = \mathrm{Tr}(A_0 \cdots A_{2k+1}) \cdot \underbrace{\left[\sum_{i=0}^{2k}(-1)^i + (-1)^{2k+1}\right]}_{= 1 + (-1) = 0} = 0. \qquad \checkmark$$

* **Schritt B: Zyklische Invarianz**
$$(\lambda\tau_{2k})(A_0,\dots,A_{2k}) = (-1)^{2k}\,\mathrm{Tr}(A_{2k} A_0 A_1 \cdots A_{2k-1}) = \mathrm{Tr}(A_0 \cdots A_{2k}) = \tau_{2k}(A_0,\dots,A_{2k}). \qquad \checkmark$$

* (Hier nutzt man wieder Spur-Zyklizität.)

* **Folgerung**:
$$[\tau_{2k}] \in HC^{2k}(M_n(\mathbb{C})), \qquad k = 0, 1, 2, \dots$$

* Diese Klassen werden durch den Periodizitäts-Operator $S$ aufeinander abgebildet:
$$S[\tau_{2k}] = [\tau_{2k+2}] \quad \text{(bis auf nicht-verschwindenden Vorfaktor)}.$$

6. Morita-Invarianz und der Schluss

* Es gilt das **Morita-Theorem für zyklische Kohomologie**:
$$HC^*(M_n(A)) \cong HC^*(A) \quad \text{für jede Algebra } A.$$
* Mit $A = \mathbb{C}$ folgt $HC^*(M_n(\mathbb{C})) = HC^*(\mathbb{C}) = (\mathbb{C}, 0)$-periodisch — und wir haben gerade explizit den Erzeuger in jedem Grad konstruiert.
* Die explizite Isomorphie wird durch die *verallgemeinerte Spur* realisiert: für $\varphi \in C^*(\mathbb{C})$ sei
$$\Phi(A_0, A_1, \dots, A_n) := \varphi(1, \dots, 1) \cdot \mathrm{Tr}(A_0 A_1 \cdots A_n).$$
* Das ist die Hauptpointe von Morita-Invarianz in der zyklischen Welt: **die Spur ist die universelle Verbindung zwischen $\mathbb{C}$ und $M_n(\mathbb{C})$**.
